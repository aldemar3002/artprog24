```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
df = pd.read_csv("housing.csv")

# Mostrar las primeras filas del dataframe
print(df.head())
```

```python


# Realizar el anlisis de variables
# Diagramas de cajas y bigotes
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Diagrama de cajas y bigotes de todas las variables")
plt.show()

```

```python


# Histogramas
plt.figure(figsize=(12, 8))
df.hist(bins=20, figsize=(12, 10))
plt.suptitle("Histogramas de todas las variables")
plt.show()

```

```python


# Mapa de calor de correlacin
# Eliminar columnas no numricas si es necesario
df_numeric = df.select_dtypes(include=['number'])
```

```python


# Calcular la correlacin entre las variables numricas
plt.figure(figsize=(12, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Mapa de calor de correlacin entre variables numricas")
plt.show()
```
