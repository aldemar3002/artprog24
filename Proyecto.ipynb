```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
df = pd.read_csv("housing.csv")

# Mostrar las primeras filas del dataframe
print(df.head())
```

```python


# Realizar el análisis de variables
# Diagramas de cajas y bigotes
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Diagrama de cajas y bigotes de todas las variables")
plt.show()

```

```python


# Histogramas
plt.figure(figsize=(12, 8))
df.hist(bins=20, figsize=(12, 10))
plt.suptitle("Histogramas de todas las variables")
plt.show()

```

```python


# Mapa de calor de correlación
# Eliminar columnas no numéricas si es necesario
df_numeric = df.select_dtypes(include=['number'])
```

```python


# Calcular la correlación entre las variables numéricas
plt.figure(figsize=(12, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Mapa de calor de correlación entre variables numéricas")
plt.show()
```
